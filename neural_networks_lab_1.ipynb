Задача №1 - 3 вариант

image.png

Синаптические коэффициенты:
$$
w_4 = w_5 = -1,
$$
$$
v_1 = v_2 = v_3 = w_1 = w_2 = w_3 = 1
$$

Смещение всех нейронов равно нулю.
На вход подается обучающий пример $$ x = 1 $$
Желаемое значение выхода $$ σ = -2 $$

## 1) Вычисляем выходы нейронов первого уровня $y_1$, $y_2$ и $y_3$:
Сначала вычисляем значения на скрытых нейронах:
$$ y_1 = x * v_1 = 1*1 = 1 $$
$$ y_2 = x * v_2 + y_1 * w_4 + y_3 * w_5 = 1*1 + 1*(-1) + 1*(-1) = -1 $$
$$ y_3 = x * v_3 = 1*1 = 1 $$

## 2) Вычисляем значения на выходных нейронах:
$$ y = y_1*w_1 + y_2*w_2 + y_3*w_3 = 1+(-1)+1 = 1 $$

Так как функция активации данных нейронов линейная, значения остаются без изменений.

## 3) Вычисление функции потерь $L$:
Функция потерь задается как:
$$ L = \frac{1}{2}(σ-y)^2 $$

Подставляем желаемое значение выхода $ σ=-2 $

$$ L = \frac{1}{2}(-2-1)^2=4,5 $$

## 4) Вычисление градиента функции потерь $\frac{\partial L}{\partial v_1}$:
### 1. Градиент по выходу $y$:
$$ \frac{\partial L}{\partial y} = -σ+y = 2+1=3 $$
### 2. Градиент выхода по активации $h: \frac{\partial y}{\partial h}$ — так как активация выходного нейрона линейная, мы можем использовать:
$$ \frac{\partial y}{\partial h} = 1 $$
### 3. Градиент по $v_1$: Мы знаем, что:
$$ \frac{\partial h}{\partial v_1} = x = 1 $$
Тогда градиент функции потерь по $v_1$ вычисляется как:
$$ \frac{\partial L}{\partial v_1} = \frac{\partial L}{\partial y} \cdot \frac{\partial y}{\partial h} \cdot \frac{\partial h}{\partial v_1} = 3*1*1 = 3$$

## 5) Обновление значения $v_1$:
Теперь, используя скорость обучения $α=0,1$:
$$ v_1^{\text{new}} = v_1^{\text{old}} - \alpha \cdot \frac{\partial L}{\partial v_1} = 1 - 0,1*3 = 0,7$$

После одной итерации обучения с учетом заданных значений, новое значение синаптического коэффициента $v_1^{\text{new}}$ равно 0.7


```python
# Задаем параметры
v1 = 1
v2 = 1
v3 = 1
w1 = 1
w2 = 1
w3 = 1
w4 = -1
w5 = -1
alpha = 0.1
sigma = -2
x = 1

# Вычисляем значения y1, y2 и y3
y1 = x * v1
y3 = x * v3
y2 = x * v2 + y1 * w4 + y3 * w5

# Вычисляем выход сети y
y = y1 * w1 + y2 * w2 + y3 * w3

# Функция потерь
L = 0.5 * (sigma * y)

# Вычисляем градиент функции потерь по y
dL_dy = -sigma + y

# Вычисляем градиент y по h
dy_dh = 1

# Градиент функции потерь по v1
dh_dv1 = x
dL_dv1 = dL_dy * dy_dh * dh_dv1

# Обновляем значение v1
v1_new = v1 - alpha * dL_dv1

print(f"Новое значение синаптического коэффициента v1: {v1_new}")
```

Синаптические коэффициенты:

На вход подается $x = 0,5$

## 1) Вычисление активаций нейронов
Для первого нейрона: $h_1 = v_1*x = 0,5$

Для второго нейрона: $h_2 = v_2*x = -0,5$

Для выходного нейрона: $h = h_1*w_1 + h_2*w_2 = 0,5 + 0,25 = 0,75$

## 2) Вероятности выхода нейроной
1. Для первого нейрона $h_1 = 0.5$, следовательно применим $0.5(h+1)$:
$$ p(h_1) = 0,5*(0,5+1) = 0,75 $$
$$ P[Y = 0] = 1 - p(h_1) = 0,25 $$
$$ P[Y = 1] = p(h_1) = 0,75 $$
2. Для второго нейрона $h_2 = -0.5$, следовательно применим $0.5(h+1)$:
$$ p(h_1) = 0,5*(-0,5+1) = 0,25 $$
$$ P[Y = 0] = 1 - p(h_1) = 0,75 $$
$$ P[Y = 1] = p(h_1) = 0,25 $$
3. Для выходного нейрона $h = 0.75$, следовательно применим $0.5(h+1)$:
$$ p(h_1) = 0,5*(0,75+1) = 0,875 $$
$$ P[Y = 0] = 1 - p(h_1) = 0,125 $$
$$ P[Y = 1] = p(h_1) = 0,875 $$

## 3) Математическое ожидание выхода сети
$$ E[Y] = 2p(h) - 1 $$
1. Для первого нейрона:
$$ E[Y_1] = 2*0,75 - 1 = 0,5 $$
2. Для второго нейрона:
$$ E[Y_2] = 2*0,25 - 1 = -0,5 $$
3. Для выходного нейрона:
$$ E[Y] = 2*0,875 - 1 = 0,75 $$

Математическое ожидание выхода сети = 0,75


```python
# Задаем параметры
x = 0.5
v1 = 1
v2 = -1
w1 = 1
w2 = -0.5

# Функция для вычисления активации нейронов
def p(h):
    if h <= -1:
        return 0
    elif -1 < h <= 1:
        return 0.5 * (h + 1)
    else:
        return 1

# Вычисление активаций нейронов
h1 = x * v1
h2 = x * v2
h = h1 * w1 + h2 * w2

# Применение функции активации для каждого нейрона
p_h1 = p(h1)
p_h2 = p(h2)
p_h = p(h)

# Вывод вероятностей выхода
def P(y, h):
  if y == 0:
    return 1 - p(h)
  else:
    return p(h)


P_Y0_h1 = P(0, h1)
P_Y1_h1 = P(1, h1)

P_Y0_h2 = P(0, h2)
P_Y1_h2 = P(1, h2)

P_Y0_h = P(0, h)
P_Y1_h = P(1, h)

# Вычисление математического ожидания выхода сети
E_Y1 = 2 * p(h1) - 1
E_Y2 = 2 * p(h2) - 1
E_Y = 2 * p(h) - 1

print(f"Новое значение синаптического коэффициента E[Y]: {E_Y}")
```

Синаптические коэффициенты:
$$
v_{12} = v_{21} = w_1 = w_2 = 1,
$$
$$
v_{11} = v_{22} = 2
$$

Смещение всех нейронов равно нулю.
На вход подается обучающий пример $$ x_1 = 1 $$ $$ x_2 = -1 $$
Желаемое значение выхода $$ σ = 3 $$

## 1) Вычисляем выходы нейронов первого уровня $y_1$ и $y_2$:
Сначала вычисляем значения на скрытых нейронах:
$$ h_1 = x_1 * v_{11} + x_2 * v_{12} = 1*2 + (-1)*1= 1 $$
$$ y_1 = h_1 $$, так как $h_1 >= 0$
$$ h_2 = x_2 * v_{22} + x_1 * v_{21} = (-1)*2 + 1*1 = -1 $$
$$ y_2 = -h_2 $$, так как $h_2 < 0$

## 2) Вычисляем значения на выходных нейронах:
$$ y = y_1 * w_1 + y_2 * w_2 = 1*1 + 1*1 = 2 $$

Так как функция активации данных нейронов линейная, значения остаются без изменений.

## 3) Вычисление функции потерь $L$:
Функция потерь задается как:
$$ L = \frac{1}{2}(σ-y)^2 $$

Подставляем желаемое значение выхода $ σ=3 $

$$ L = \frac{1}{2}(3-2)^2=0,5 $$

## 4) Вычисление градиента функции потерь $\frac{\partial L}{\partial v_1}$:
### 1. Градиент по выходу $y_1$:
$$ \frac{\partial L}{\partial y_1} = -σ+y = -3+2=-1 $$
Градиент по весам $w_1$ и $w_2$:
$$ \frac{\partial L}{\partial w_1} = \frac{\partial L}{\partial y} ⋅ y_1 =-1 $$
$$ \frac{\partial L}{\partial w_2} = \frac{\partial L}{\partial y} ⋅ y_2 =-1 $$
### 2. Градиент выхода по активации $h_1: \frac{\partial y_1}{\partial h_1}$ — так как активация выходного нейрона линейная, мы можем использовать:
$$ \frac{\partial y_1}{\partial h_1} = 1 $$
### 3. Градиент по $v_{11}$: Мы знаем, что:
$$ \frac{\partial h_1}{\partial v_11} = x_1 = 1 $$
Тогда градиент функции потерь по $v_1$ вычисляется как:
$$ \frac{\partial L}{\partial v_{11}} = \frac{\partial L}{\partial y_1} \cdot \frac{\partial y_1}{\partial h_1} \cdot \frac{\partial h_1}{\partial v_{11}} = -1*1*1 = -1$$

## 5) Обновление значения $v_1$:
Теперь, используя скорость обучения $α=0,1$:
$$ v_{11}^{\text{new}} = v_{11}^{\text{old}} - \alpha \cdot \frac{\partial L}{\partial v_{11}} = 2 - 0,1*(-1) = 2,1$$

После одной итерации обучения с учетом заданных значений, новое значение синаптического коэффициента $v_{11}^{\text{new}}$ равно 2,1

```python
# Задаем параметры
v11 = 2
v12 = 1
v21 = 1
v22 = 2
w1 = 1
w2 = 1
alpha = 0.1
sigma = 3
x1 = 1
x2 = -1

# Вычисляем значения y1, y2
h1 = x1 * v11 + x2 * v12
h2 = x2 * v22 + x1 * v21

def y(h):
  if h >= 0:
    return h
  else:
    return -h

y1 = y(h1)
y2 = y(h2)

# Вычисляем выход сети y
y = y1 * w1 + y2 * w2

# Функция потерь
L = 0.5 * (sigma * y)

# Вычисляем градиент функции потерь по y
dL_dy1 = -sigma + y

# Вычисляем градиент y по h
dy1_dh1 = 1

# Градиент функции потерь по v1
dh_dv11 = x1
dL_dv11 = dL_dy1 * dy1_dh1 * dh_dv11

# Обновляем значение v1
v11_new = v11 - alpha * dL_dv11

print(f"Новое значение синаптического коэффициента v1: {v11_new}")
```
